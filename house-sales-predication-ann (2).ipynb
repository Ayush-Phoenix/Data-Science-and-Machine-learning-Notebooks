```python
df.drop(['id'],axis =1 ,inplace=True)
```

```python
df['date'] =pd.to_datetime(df['date'])
```

```python
df
```

```python
df['year']= df['date'].apply(lambda date: date.year)
df['month']= df['date'].apply(lambda date: date.month)
df.drop(['date'],axis=1, inplace=True)
```

```python
df.drop(['zipcode'],axis=1, inplace=True)
```

```python
x = df.drop(['price'],axis=1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,train_size=0.3,random_state=69)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
DENSE_LAYERS = 19
model = Sequential()
model.add(Dense(DENSE_LAYERS,activation='relu'))
model.add(Dense(DENSE_LAYERS,activation='relu'))
model.add(Dense(DENSE_LAYERS,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
EPOCH = 200
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size = 128,epochs = EPOCH) 
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
 pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test,pred)
mae
```
