# Importing required libraries 

```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

Loding data from sklearn libraries

```python
cancer = load_breast_cancer()
cancer.keys()
```

# reading about data set, to get better insight 

```python
print(cancer['DESCR'])
```

```python
# creadting DataFrame from the cancer data and its keys
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
# chaacking for missing data
df.isna().sum()
```

# Preparing and spliting data for training and testing

```python
x = df
y = cancer['target']
```

```python
x_train , x_test, y_train , y_test = train_test_split(x,y,test_size=0.3, random_state=69)
```

# Training the model 

```python
model = SVC()
model.fit(x_train,y_train)
pred = model.predict(x_test)
```

```python
#cheacking accuracy 
accuracy_score(pred,y_test)
```
